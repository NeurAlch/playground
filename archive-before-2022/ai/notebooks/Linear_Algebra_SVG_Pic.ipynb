```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
pic = Image.open('Einstein_tongue.jpg')

plt.imshow(pic)
plt.show()
```

```python
pic = np.array(pic)
plt.imshow(pic, cmap='gray')
plt.show()
```

```python
U, S, V = np.linalg.svd(pic)
```

```python
plt.plot(S, 's-')
plt.xlim(([0, 400]))
plt.xlabel('Component number')
plt.ylabel('Singular value (\sigma)')
plt.show()
```

```python
comps = np.arange(1, 2)
comps
```

```python
# low rank approximation version of the pic
reconPic = U[:,comps] @ np.diag(S[comps]) @ V[comps,:]
```

```python
plt.figure(figsize=(5, 10))
plt.subplot(1, 2, 1)
plt.imshow(pic, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(reconPic, cmap='gray')
plt.title('Component %s-%s' % (comps[0], comps[-1]))
plt.axis('off')

plt.show()
```
